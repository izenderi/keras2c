```python
import matplotlib.pyplot as plt

def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='latin1')
    return dict

file = r'/home/johnson/dataset/cifar-10-batches-py/data_batch_1'
data_batch_1 = unpickle(file)
# print(type(data_batch_1))
# print(data_batch_1.keys())

meta_file = r'/home/johnson/dataset/cifar-10-batches-py/batches.meta'
meta_data = unpickle(meta_file)
# print(type(meta_data))
# print(meta_data.keys()) 

# image = data_batch_1[b'data'][0]
# image = image.reshape(3,32,32)
# print(image.shape)
# label names
label_name = meta_data['label_names']
# take first image
image = data_batch_1['data'][3]
# take first image label index
label = data_batch_1['labels'][3]
# Reshape the image
image = image.reshape(3,32,32)
# Transpose the image
image = image.transpose(1,2,0)
# Display the image
plt.imshow(image)
plt.title(label_name[label])
```
